```python
# Import Librabry 
import numpy as np
import csv
```

```python
# read the data with csv into list 'df'
df = []
csvfile = open("./data.csv", "r")
reader_variable = csv.reader(csvfile, delimiter=",")
for data in reader_variable:
   df.append(data)


for data in range(0,5):
    print(df[data])
    

```

```python
df = np.array(df[1:])
print(df.shape)
```

```python
# features name
df[:1]
```

```python
# Find the min and max values for each column
def dataset_minmax(df):
    minmax = list()
    for i in range(len(df[0])-1):
        col_values = [row[i] for row in df]
        value_min = min(col_values)
        value_max = max(col_values)
        minmax.append([value_min, value_max])
    return minmax

# normalize
minmax = dataset_minmax(df)
print(minmax)
```
