```python
import pandas as pd
import datetime
```

- Firstly, we will join the customers and orders table using an INNER JOIN with join condition being “id” column of customers table is equal to “cust_id” column of orders table.
- Then, we will filter the date for dates between 2019-02-01 and 2019-05-01.
- Then, we will group the data on the basis or customer first_name and order_date.
- Then we will output the customer first_name, order_date and SUM(total_order_cost). This output will be displayed on the basis of the GROUP BY condition that we have created in step 3. We are using SUM() function because a customer can order multiple times in a single day. So SUM() will sum all the transaction values for that day for the customer.
- Next step is to order the date based on SUM(total_order_cost) in Descending Order.
- Last step is to use “LIMIT 1” to display only the employee with the highest order cost.

```python
customers = pd.read_csv('customers.csv')
orders = pd.read_csv('orders.csv')
```

```python
Both_DFs = pd.merge(customers,orders, how='inner',left_on=['id'],right_on=['cust_id'])
```

```python
Both_DFs.info()
```

```python
final = Both_DFs[(Both_DFs['order_date']>"2019-02-01 0:00:00") & (Both_DFs['order_date']<"2019-05-01 0:00:00")]
```

```python
final
```

```python
final1 = final.groupby(['first_name','order_date'])['total_order_cost'].sum()
```
