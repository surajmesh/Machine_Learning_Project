Performing Uni-variate Data Analysis

```python
# Using Box plot let visulized the distribution of Data and oberserve Outlier in each variable
```

```python
import plotly.graph_objects as go



fig = go.Figure()
# Use x instead of y argument for horizontal plot
fig.add_trace(go.Box(y=data.sulphates))
fig.add_trace(go.Box(y=data.chlorides))



fig.show()
```

```python
data.columns
```

```python
# initialize figure with 8 subplots in a row
fig, ax = plt.subplots(1, 8, figsize=(15, 6))

# add padding between the subplots
plt.subplots_adjust(wspace=0.7) 


sns.boxplot(data=data['fixed acidity'], ax=ax[0], color='brown',)
ax[0].set_xlabel('fixed acidity')


sns.boxplot(data=data['volatile acidity'], ax=ax[1], color='g')
ax[1].set_xlabel('volatile acidity')


sns.boxplot(data=data['citric acid'], ax=ax[2], color='y')
ax[2].set_xlabel('stores_count')


sns.boxplot(data=data['residual sugar'], ax=ax[3])
ax[3].set_xlabel('residual sugar')

sns.boxplot(data=data['free sulfur dioxide'], ax=ax[4], color='brown',)
ax[4].set_xlabel('free sulfur dioxide')


sns.boxplot(data=data['total sulfur dioxide'], ax=ax[5], color='g')
ax[5].set_xlabel('total sulfur dioxide')


sns.boxplot(data=data['density'], ax=ax[6], color='y')
ax[6].set_xlabel('density')


sns.boxplot(data=data['alcohol'], ax=ax[7])
ax[7].set_xlabel('alcohol')

for subplot in ax:
    subplot.set_xticklabels([])
    
plt.show()
```
