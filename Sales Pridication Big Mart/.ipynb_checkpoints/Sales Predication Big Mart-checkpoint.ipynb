```python
#create figurewith 1*2 subplot
fig,axes = plt.subplots(1,2,figsize=(12,8)) 

sns.distplot(train_df["Item_Weight"],rug=True,hist=False,ax=axes[0],rug_kws={"color": "red"})
axes[0].set_title("Item Weights Distribution")

sns.distplot(train_df["Item_Visibility"],rug=True,hist=False,ax=axes[1],rug_kws={"color": "red"})
axes[1].set_title("Item Visibility Distribution")
```

```python
#create figurewith 1*2 subplot
fig,axes = plt.subplots(1,2,figsize=(12,8)) 

sns.distplot(train_df["Item_MRP"],rug=True,hist=False,ax=axes[0],rug_kws={"color": "red"})
axes[0].set_title("Item MRP Distribution")

sns.distplot(train_df["Outlet_Establishment_Year"],rug=True,hist=False,ax=axes[1],rug_kws={"color": "red"})
axes[1].set_title("Outlet_Establishment_Year Distribution")
```

```python
# Now let detect outlier in variable by Visulizing Box Plot
```

```python
# initialize figure with 4 subplots in a row
fig, ax = plt.subplots(1, 4, figsize=(15, 6))

# add padding between the subplots
plt.subplots_adjust(wspace=0.7) 


sns.boxplot(train_df["Item_Weight"], ax=ax[0], color='brown',)
ax[0].set_xlabel('Item Weights Distribution')


sns.boxplot(train_df["Item_Visibility"], ax=ax[1], color='g')
ax[1].set_xlabel('Item Visibility Distribution')


sns.boxplot(train_df["Item_MRP"], ax=ax[2], color='y')
ax[2].set_xlabel('Item MRP Distribution')


sns.boxplot(train_df["Outlet_Establishment_Year"], ax=ax[3])
ax[3].set_xlabel('Outlet_Establishment_Year Distribution')





for subplot in ax:
    subplot.set_xticklabels([])
    
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(train_df.corr(),cmap="gray_r",annot=True, fmt=".2f")
```

```python
#only display cells where the correlation value is greater than 0.5 using heatmap
plt.figure(figsize=(10,5))

corr_matrix = train_df.corr()
mask = (corr_matrix.abs() > 0.5)
sns.heatmap(corr_matrix[mask],cmap="gray_r",annot=True, fmt=".2f",vmin=-1, vmax=1)
```
